## Investigation Overview



In this investigation, I wanted to look the sales of different branches, payment mode, gender preferences for the total sales in the three brances of the superstores located in 3 different cities.

## Dataset Overview

The data consisted of Sales figures of three superstores located in different cities. The total records were 1000.
The attributes included Gender,City,Branch,Total. The data was clean and there was no need to wrangle the data.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
sales=pd.read_csv('supermarket_sales.csv',parse_dates=['Date'])
```

## Total Sales distribution for number of Orders

Total has a long tailed distribution with a lot of total sales are at low total ends and few on thee high total end.

```python
plt.figure(figsize=[8, 5])

plt.hist(sales['Total'],bins=10)
plt.xlabel('Total Sales')
plt.ylabel('No of Orders')
plt.title('Total Sales with their count')
;
```

## Genderwise puchase of Product line

More orders in Fashion accessories by Female than men. Health and beauty has more male orders.
Sports and travel have Female. Are female spending more on sports than men or they are travelling more than men?
This is the question we need to ask. Do female shoppers are shopping for sports item for their kids?. These questions 
can be posed.

```python
plt.figure(figsize=(12, 6))
plt.title('Total Monthly transaction')
sb.countplot(sales['Product line'], hue = sales['Gender'])
;
```

## Payment method comparision for number of purchases in each branch genderwise

 1: Ewallet is highest used in Branch A, While Cash is highest used in Branch C.
Question is why Branch C has highest Cash users? also they are the least credit card users. Any reason?
What is the income levels of Branch C customers? Is it industrial area with less education and labor intensive?
Observation 2: More Female shoppers in Branch C then men. Also Branch C has more Cash transactions
then men. So we can conclude in Branch C, there are more female shoppers and their payment preference is Cash.
Question is are the women folks located in Branch C do jobs where they get money in cash daily?

```python
plt.figure(figsize=[16,5])
plt.subplot(1,2,1)
sb.countplot(data=sales,x='Branch',hue='Payment')
plt.subplot(1,2,2)
sb.countplot(data=sales,x='Branch',hue='Gender');
```

## Buying pattern according to the days of the month
Initial 10 days(Salary days?) sees an upticks followed by some decrease. Before tapering out after 27th day, it shows significant increase.

```python
from matplotlib.pyplot import figure

sales['day']=pd.to_datetime(sales['Date']).dt.day
df_sales=sales['day'].value_counts()
dy=sales['day'].value_counts().index
figure(figsize=(10, 8))

plt.bar(dy,df_sales)
plt.xlabel('Day of the month')
plt.ylabel('Purchases')
plt.title('Daywise Purchase count')
plt.show();
```

## Explore relationship between rating and Product line
Highest average ratings belong to Food and beverages. Do these supermarket sells high quality of food and beverages?
Home and lifestyle has the lowest average ratings. Low Quality goods may be the reason.

```python
figure(figsize=(10, 8))
base_color=sb.color_palette()[0]

sb.pointplot(data=sales,x='Product line',y='Rating',color=base_color,linestyles="")
plt.xticks(rotation=90);
```
